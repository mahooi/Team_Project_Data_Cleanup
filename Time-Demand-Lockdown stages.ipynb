```python
#Figure one: Lockdown Vs. Demand Evolution

%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("/Users/ibsi/Desktop/Ironhack/butter_team_project/sql_merged_data/sh_dem_2020_no_dup.csv")
```

```python
df=pd.DataFrame(data)
df
```

```python
Demand_plot = df.loc[data["Country"] == "Spain"]
fig, ax = plt.subplots(figsize = (10,8))
ax.plot(Demand_plot["Date"], Demand_plot["Value"])
ax.set_title("Electricity demand before and during COVID-19 lockdown")
ax.set_xlabel('Date')
ax.set_ylabel('Value')
```
