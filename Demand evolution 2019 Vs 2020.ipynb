Figure 2: Demand Evolution 2019 and 2020

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("/Users/ibsi/Desktop/Ironhack/butter_team_project/sql_merged_data/total_dem.csv")
```

```python
df=pd.DataFrame(data)
df
```

```python
Demand_2019 = df.iloc[0:165]
Demand_2020 = df.iloc[166:332]
fig, ax = plt.subplots(figsize = (30,8))
ax2 = ax.twiny()
ax.set_title("Electricity demand in 2019 and 2020")

Demand_2019_smoothed=Demand_2019.rolling(7).sum()
Demand_2019_smoothed["Date"]=Demand_2019["Date"]

Demand_2020_smoothed=Demand_2020.rolling(7).sum()
Demand_2020_smoothed["Date"]=Demand_2020["Date"]

Demand_2019_smoothed.plot(kind='line',x='Date',y='Value',ax=ax, color = "red")
Demand_2020_smoothed.plot(kind='line',x='Date',y='Value',ax=ax2)
```
