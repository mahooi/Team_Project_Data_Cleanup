```python
import pandas as pd
```

```python
url = '../butter_team_project/initial_db/stay-at-home-covid.csv'

requirements = pd.read_csv(url, sep = ',', index_col = False)
```

```python
requirements.head()
```

```python
# Create a dictionary for the requirements
req_description = {0: "No measures", 1: "Recommended not to leave the house", 2: "Required to not leave the house with exceptions for daily exercise, grocery shopping, and ‘essential’ trips", 3: "Required to not leave the house with minimal exceptions (e.g. allowed to leave only once every few days, or only one person can leave at a time, etc."}

# Create a new column and populate it with values in the Stay at home requirements. 
requirements['Description'] = requirements['Stay at home requirements (OxBSG)'].map(req_description)
requirements.head()
```

```python
def get_data_country(country):
    by_country = requirements['Entity']==country
    return requirements[by_country]

x = get_data_country("Spain")
```

```python
# Rename the column "Stay at home requirements (OxBSG)" (original)
requirements = requirements.rename({'Stay at home requirements (OxBSG)': 'LockdownMeasuresId'}, axis=1)
requirements.head()
```
